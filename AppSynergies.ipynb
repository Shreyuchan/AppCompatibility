```python
#First we will create 4 user profiles (male and female) and ask them the onboarding questions
#Their answers will be chosen at random 
```

```python
import pandas as pd
import numpy as np
import random 
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn import preprocessing
```

```python
names = []
gender = []
#Loop to take in only 4 names
for i in range(0, 4):
    #Take in users name
    print("Enter your full name")
    n = str(input())
    names.append(n)
    #Ask them to enter their gender
    print("Enter your gender, whether Male or Female")
    g = str(input())
    gender.append(g)
```

```python
names
```

```python
gender
```

```python
df = pd.DataFrame(columns = ["Name" , "Gender"])
df["Name"] = names
df["Gender"] = gender
```

```python
#2 unique profiles
df["Gender"].nunique()
```

```python
questions = ["Q1. When it comes to love , Option 1 - Opposite attracts, Option 2- Similarities make things smoother",
             "Q2. When comes to love, Option 1 If it’s right, it’s easy, Option 2 - It always requires work",
             "Q3. When it comes to relationships, Option 1 - Communication is key, Option 2 - It’s complicated",
             "Q4. Predictability always brings comfort, Option 1 - Yes, Option 2 - No",
             "Q5. Which is more important?, Option 1 - Work life balance, Option 2 - Success", 
             "Q6. When trust is broken? Option 1 - The relationship is over, Option 2 - There might be a chance",  
             "Q7. My close friend and family would say, Option 1 - I’m a social butterfly, Option 2 - I’m bit reserved or even shy", 
             "Q8. My close friend and family would say, Option 1- I’m an open book, Option 2 - It takes time to know about me ", 
             "Q9. My close friend and family would say, Option 1 - I’m opinionated and strong-willed , Option 2 - I’m easier going and laid-back", 
             "Q10. Having common interests , Option 1 - Can make or break a relationship, Option 2 -Doesn’t matter much"]
```

```python
len(questions)
```

```python
#Develop the database for the compatibility questions
#Answers are taken at random
compatibility_questions = []
choice = [0, 1]
answer = []
for i in df["Name"]:
    #Taking all the answers from one candidate for one questions
    print("Hello", i)
    print("Welcome to the Teletubby app")
    print("Please answer the questions below, select 0 for option 1 and 1 for option 2")
    for j in questions:
        print(j)
        a = random.choice(choice)
        answer.append(a)
    compatibility_questions.append(answer)
    answer = []
```

```python
len(compatibility_questions)
```

```python
for i in compatibility_questions:
    print(len(i))
    print(i)
    break
```

```python
df["Answers"] = compatibility_questions
df
```

```python
df1 = df.explode("Answers")
df1
```

```python
df2 = pd.DataFrame(df['Answers'].to_list(), columns=['answers1','answers2','answers3','answers4','answers5','answers6',
                                                     'answers7','answers8','answers9','answers10' ])
df2
```

```python
df3 = df.iloc[:, 0:2]
df3
```

```python
final_df = pd.concat([df3, df2], sort=True, axis = 1)
final_df
```

```python
#Encoding the labels
le = preprocessing.LabelEncoder()
```

```python
le.fit(final_df["Name"])
final_df["Encoded_names"] = le.fit_transform(final_df["Name"])
```

```python
final_df
```

```python
final_df.columns
```

```python
final_df_feed = final_df.iloc[:,1:]
final_df_feed
```

```python
#Encoding the gender
le.fit(final_df_feed["Gender"])
final_df_feed["Gender_encoded"] = le.fit_transform(final_df_feed["Gender"])
```

```python
final_df_feed
```

```python
final_cluster_fed = final_df_feed.iloc[:, 1:]
final_cluster_fed
```

```python
final_cluster_fed.columns
```

```python
#Using Agglomerative clusters to cluster based on similarity of answers
# Instantiating HAC
hac = AgglomerativeClustering(n_clusters=2)

# Fitting
hac.fit(final_cluster_fed)

# Getting cluster assignments
cluster_assignments = hac.labels_

# Assigning the clusters to each profile
final_df['Cluster'] = cluster_assignments

# Viewing the dating profiles with cluster assignments
final_df
#Match between brad pitt, angelina and marie cutie
```
